```python
import pandas as pd
```

```python
df=pd.read_csv("owid-covid-data.csv")
```

```python
# check 
df.columns
```

```python
# Preview 
df.head(50)
```

```python
# identify missing values
df.isnull().sum()
```

```python
# cleaning data
df['date'] = pd.to_datetime(df['date'])
```

```python
df['total_cases'].fillna(method='ffill', inplace=True)
```

```python
df = df.dropna(subset=['date', 'total_cases', 'total_deaths'])
```

```python
df.head(100)
```

```python
# selecting relevant columns
df = df[['date', 'location', 'total_cases', 'new_cases', 'total_deaths', 'new_deaths', 'total_vaccinations']]
```

```python
df.head()
```

```python
# countries of interest
countries = ['Kenya', 'United States', 'India']
df = df[df['location'].isin(countries)]
```

```python
df['location'].head(10)
```

```python
# Drop the missing values beacuse this are critical ines
df = df.dropna(subset=['date', 'location'])
```

```python
df.head(50)
```

```python
# handling a missing value
df[['total_cases', 'new_cases', 'total_deaths', 'new_deaths', 'total_vaccinations']] = \
    df[['total_cases', 'new_cases', 'total_deaths', 'new_deaths', 'total_vaccinations']].interpolate()

df = df.fillna(0)
```

```python
df.head(50)
```

```python
# confirming no missing values
print(df.isnull().sum())
print(df.head())
```

```python
# Total cases over time
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
for country in df['location'].unique():
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['total_cases'], label=country)

plt.title('Total COVID-19 Cases Over Time')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Total Deaths Over Time
plt.figure(figsize=(10, 6))
for country in df['location'].unique():
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['total_deaths'], label=country)

plt.title('Total COVID-19 Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Daily new comparisons
plt.figure(figsize=(10, 6))
for country in df['location'].unique():
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['new_cases'], label=country)

plt.title('Daily New COVID-19 Cases')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Death rate
df['death_rate'] = df['total_deaths'] / df['total_cases']

plt.figure(figsize=(10, 6))
for country in df['location'].unique():
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['death_rate'], label=country)

plt.title('COVID-19 Death Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Death Rate')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Visualizing Vaccination progress
# cumulative vaccinations over time
plt.figure(figsize=(10, 6))
for country in df['location'].unique():
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['total_vaccinations'], label=country)

plt.title('COVID-19 Total Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Pie chart of vaccinated and Unvaccinated
latest = df[df['location'] == 'Kenya'].sort_values('date').iloc[-1]
vaccinated = latest['total_vaccinations']
population = latest.get('population', 1)  # Avoid ZeroDivisionError

unvaccinated = max(population - vaccinated, 0)

plt.figure(figsize=(6, 6))
plt.pie([vaccinated, unvaccinated],
        labels=['Vaccinated', 'Unvaccinated'],autopct='%1.1f%%')
plt.title('Kenya Vaccination Coverage (Latest Available)')
plt.show()
```

```python
#Insights
 Insights:
Vaccination Rollout Speeds Vary Drastically

The United States began mass vaccinations earlier and at a faster rate compared to Kenya and India.

India Experienced the Most Pronounced Second Wave

A sharp increase in new cases was seen in mid-2021, followed by a high death toll.

Death Rates Decreased Over Time

Despite rising total cases, the death rate declined as vaccinations increased and treatments improved.

Kenya Had Slower Vaccination Uptake

Daily New Cases Show Seasonal or Wave Patterns

Each country experienced clear waves of rising and falling case counts, often synchronized globally.

```
